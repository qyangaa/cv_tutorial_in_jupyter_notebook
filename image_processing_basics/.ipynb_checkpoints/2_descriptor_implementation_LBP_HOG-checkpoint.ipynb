# Local Binary Patterns (LBP)
<a href='https://www.pyimagesearch.com/2015/12/07/local-binary-patterns-with-python-opencv/'> Reference </a>
1. Definition: compute local representation of texture, by comparing each pixel with its surrounding neighborhood of pixels.
2. Steps
    1. Convert image to grayscale
    2. Select a neighborhood of size r surrounding the center pixel
    3. Calculate LBP value for the center pixel and store output in 2D array with the same width and height as input image
   <img src="https://www.pyimagesearch.com/wp-content/uploads/2015/12/lbp_calculation-1024x299.jpg">
        1. If neighbor >= center: set value to 1;else 0
        2. go around the neighbor to construct an array of binaries
        3. convert to decimal
        4. Store the number as LBP for the pixel corresponding to center location in output
    4. Compute a histogram over the output LBP array
3. Properties
    1. To capture details at varying scales (Ojala et. al), capture neighbors in a circle, change the radius of circle
    2. LBP is considered uniform if at most two 1-0 or 0-1 transitions
    3. Not rotation invariant, need extra steps to introduce rotation invariance. <a href='https://pdfs.semanticscholar.org/3e70/ff6ad37af67c92433426da5b08bd64ecbe2e.pdf'> Reference </a> 

````python
#ref: Kaikeba
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def get_LBP(gray_image, window=3):
    lbp = np.zeros_like(gray_image)
    h,w = gray_image.shape
    p_center = window//2
    for ph in range(0,h-window):
        for pw in range(0,w-window):
            img = gray_image[ph:ph+window, pw:pw+window]
            center = img[p_center,p_center]
            binary = (img>=center)*1.0
            binary_vector = binary.flatten()
            #remove center point
            binary_vector = np.delete(binary_vector,4)
            res = 0
            for i in range(len(binary_vector)):
                if binary_vector[i]==1:
                    res+= 2**i
            lbp[ph+p_center,pw+p_center] = res
    return lbp
        
````

````python
img = cv2.imread('img.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
lbp = get_LBP(img_gray)
````

````python
hist = lbp.flatten()
fig = plt.figure(figsize = (20,8))
ax = fig.add_subplot(1,3,1)
ax.imshow(img_gray,'gray')
ax = fig.add_subplot(1,3,2)
ax.imshow(lbp,'gray')
ax = fig.add_subplot(1,3,3)
freq,lbp,_ = ax.hist(hist,bins=255)
ax.set_title('LBP histogram')
````

````python
# To obtain rotation invariance
def binary_min(binary):
    '''Take array of binary, and find the minimum rotation instance'''
    min_b = []
    for i in range(len(binary))[::-1]: #reverse
        print('binary:', binary)
        print('min_b:', min_b)
        if binary[i]==0:
            binary = binary[:i]
            min_b = min_b + [0]

        else:
            min_b = min_b + binary
            return np.array(min_b)
        
````

# HOG (Histogram of Oriented Gradients)
1. BGR2GRAY: gradient only for grey scale
    1. can involve aforementioned transformations, such as gamma transformation.
2. calculate gradient for every pixel (magnitude and direction)
    1. use gradient filters (kernel), eg. y gradient, use this kernel to convolute over whole image to get directional gradient for each pixel:
\begin{matrix}
-1 & -1 & -1\\
0 & 0 & 0\\
1 & 1 & 1\\
\end{matrix}
    2. magnitude of gradient: $\sqrt{dx^2+dy^2}$ 
    3. orientation: $\arctan{\dfrac{dy}{dx}}$
    4. output: magnitude matrix, orientation matrix
3. divide image into small cells (eg 8*8), with 8 * 8 * 2 values, 2 =(magnitude and orientation)
4. obtain gradient histogram for each cell
    1. ```plot(angle, count)```: angle in range, eg. [0,20], [20,40]....
    2. output shape: (1, # ranges)
5. group every 3*3 cells into a block in a similar fasion to convolution (with block size and stride=1). Concatenate histogram descriptors of these cells: concat([...],[...],[...]) = [.........].
    1. output: (1, 3 * 3 * #ranges)
    2. can add one more layer between cell and block (window)
6. concat descriptors of every block, normalize to give the final descriptor.


## Sample pipeline
Input image -> normalize gamma & color -> compute gradients -> weighted vote into spatial & orientation cells -> contrast normalize over overlapping spatial blocks -> collect HOG's over detection window -> Linear SVM -> classification

## A closer look at HOG Descripor

````python
def hog_descriptor(image):
    hog = cv2.HOGDescriptor()
    h, w = image.shape[:2]
    rate = 64 / w
    image = cv2.resize(image, (64, np.int(rate*h)))
    bg = np.zeros((128, 64), dtype=np.uint8)
    print(bg.shape,image.shape)
    bg[:,:] = 127
    h, w = image.shape
    dy = (128 - h) // 2
    bg[dy:h+dy,:] = image
    descriptors = hog.compute(bg, winStride=(8, 8), padding=(0, 0))
    return descriptors
````

````python
from skimage import io
url = 'http://absfreepic.com/absolutely_free_photos/small_photos/walking-people-on-street-2500x1667_84106.jpg'
img2 = io.imread(url)

img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)[50:350,0:300]
plt.imshow(img2_gray,cmap="gray")
````

````python
img2_gray.shape
````

````python
descriptor1 = hog_descriptor(img2_gray)
plt.plot(descriptor1)
````

````python
def svm_predict(feature):
    weight = cv2.HOGDescriptor_getDefaultPeopleDetector()
    result = np.sum(weight[:-1]*feature)+weight[-1]
    return result
````
